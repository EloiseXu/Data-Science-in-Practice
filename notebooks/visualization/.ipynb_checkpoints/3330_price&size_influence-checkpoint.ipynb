Ratings & revenues analysis by app size and app price

Purpose 
This notebook will find out how an app's size and price influence an app's rating and revenue. Will apps be more likely to get higher ratings or revenues in google play when they have certain genres? Will some genres help apps get higher ratings or revenues in apple store?

```python
import math
import pandas as pd
import numpy as np

from matplotlib.pylab import plt
import seaborn as sns

from os.path import exists, isfile
```

```python
if not exists("datasets/2300_combine_kaggle_datasets.csv"):
    print ("Missing dataset file")
    
df=pd.read_csv("datasets/2300_combine_kaggle_datasets.csv")
df.head()
```

```python
df = df[['normed_apple_rating', 'apple_size', 'normed_google_rating', 'google_size', 'price', 'log_google_reviews', 'google_reviews', 'apple_reviews', 'log_apple_reviews', 'genre']].copy()
df['google_business'] = df['google_reviews'] * df['price']
df['apple_business'] = df['apple_reviews'] * df['price']

df.loc[df['google_business']>0,'log_google_business'] = df[df['google_business']>0]['google_business'].apply(lambda x: math.log(x, 10))
df.loc[df['apple_business']>0,'log_apple_business'] = df[df['apple_business']>0]['apple_business'].apply(lambda x: math.log(x, 10))
```

```python
corr_df = df[['normed_apple_rating', 'apple_size', 'normed_google_rating', 'google_size', 'price', 'log_google_reviews', 'log_apple_reviews', 'log_google_business', 'log_apple_business']].corr() 

plt.figure(figsize=(6,6))
plt.title("correlation of features")
sns.heatmap(corr_df)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

ax.set_xlim(0, 100)
ax.set_ylim(0, 7.5)

sns.regplot('apple_size', 'log_apple_reviews', data=df[df['apple_size']<100], ax=ax, truncate=True, order=3, label='apple')
sns.regplot('google_size', 'log_google_reviews', data=df[df['apple_size']<100], ax=ax, truncate=True, order=3, label='google')
plt.legend()
```

```python
fig, bx = plt.subplots(figsize=(6, 6))

sns.regplot('price', 'log_apple_reviews', data=df[df['price']<=15], ax=bx, truncate=True, label='apple')
sns.regplot('price', 'log_google_reviews', data=df[df['price']<=15], ax=bx, truncate=True, label='google')
plt.legend()
```
