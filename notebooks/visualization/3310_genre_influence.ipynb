# RQ3 - Do some apps better received on Google Play while other do not?

__Purpose__

This notebook will find out how an app's genre influences an app's rating. Will apps be more likely to get higher ratings in google play when they have certain genres? Will some genres help apps get higher ratings in apple store?

```python
import math
import pandas as pd
import numpy as np

from matplotlib.pylab import plt
from matplotlib.colors import ListedColormap

from os.path import exists, isfile
```

# Load datasets

```python
if not exists("../../datasets/2100_clean_google.csv"):
    print ("Missing dataset file")
    
df_google=pd.read_csv("../../datasets/2100_clean_google.csv")
df_google.head()
```

```python
if not exists("../../datasets/2200_clean_apple.csv"):
    print ("Missing dataset file")
    
df_apple=pd.read_csv("../../datasets/2200_clean_apple.csv")
df_apple.head()
```

```python
result_dir = '../../results/'
```

# The frequency of different app genres

This chart shows the proportion of each genre in quantity.

```python
fig, ax = plt.subplots(figsize=(15, 6))

google_genre = df_google.groupby('google_genre')['google_genre'].count() / df_google['google_genre'].count()
apple_genre = df_apple.groupby('apple_genre')['apple_genre'].count() / df_apple['apple_genre'].count()
df_genre =pd.concat([google_genre, apple_genre],axis=1)

df_genre.plot(y = ['google_genre', 'apple_genre'], kind = 'bar', ax = ax)

ax.set_ylabel('Frequency')
ax.set_xlabel('Genre')
plt.xticks(rotation=(-60), fontsize=12)
```

Load dataset: common apps

```python
if not exists("../../datasets/2300_combine_kaggle_datasets.csv"):
    print ("Missing dataset file")
    
df=pd.read_csv("../../datasets/2300_combine_kaggle_datasets.csv")
df.head()
```

# Compare the normalized ratings and z score ratings of apps

```python
fig, ax = plt.subplots(figsize=(15, 6))

google_genre = df.groupby('genre')['normed_google_rating'].mean()
apple_genre = df.groupby('genre')['normed_apple_rating'].mean()
df_genre =pd.concat([google_genre, apple_genre],axis=1)

df_genre.plot(y = ['normed_google_rating', 'normed_apple_rating'], kind = 'bar', ax = ax)

ax.set_ylabel('Normalized Rating')
ax.set_xlabel('Genre')
plt.xticks(rotation=(-60), fontsize=12)
```

```python
fig, ax = plt.subplots(figsize=(15, 6))

google_genre = df.groupby('genre')['z_score_google_rating'].mean()
apple_genre = df.groupby('genre')['z_score_apple_rating'].mean()
df_genre =pd.concat([google_genre, apple_genre],axis=1)

df_genre.plot(y = ['z_score_google_rating', 'z_score_apple_rating'], kind = 'bar', ax = ax)

ax.axhline(c='k')

ax.set_ylabel('Z score')
ax.set_xlabel('Genre')
plt.xticks(rotation=(-60), fontsize=12)
```

Across the genres, the google ratings varies  more significantly than apple ratings.
